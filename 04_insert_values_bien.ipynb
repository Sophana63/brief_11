```python
import pandas as pd
from pandasql import sqldf
import mysql.connector as mysql
```

```python
df_vf = pd.read_excel('Donnees_Immo/Valeurs_foncières.xlsx')
df_vf = df_vf[['No voie', 'Type de voie', 'Voie', 'Nombre pieces principales', 'Surface Carrez du 1er lot', 'Surface reelle bati', 'Type local', 'Code departement', 'Code commune']]
df_vf["id_code"] = df_vf['Code departement'].map("{:0>2}".format) + df_vf['Code commune'].map("{:03}".format)
df_vf['id'] = df_vf.index
bien = df_vf.astype(object).where(pd.notnull(df_vf), None)
bien = bien[['id', 'No voie', 'Type de voie', 'Voie', 'Nombre pieces principales', 'Surface Carrez du 1er lot', 'Surface reelle bati', 'Type local', 'id_code']]

bien.head(238)
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "valeur_fonciere"
)

cursor = db.cursor()

sql = "INSERT INTO bien(id, num_voie, type_voie, voie, total_piece, surface_carrez, surface_local, type_local, id_com) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s)"
for i,row in bien.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
