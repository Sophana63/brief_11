```python
import pandas as pd
from pandasql import sqldf
import mysql.connector as mysql
```

```python
df_vente = pd.read_excel('Donnees_Immo/Valeurs_foncières.xlsx')
df_vente['id'] = df_vente.index
#df_ = pd.to_datetime(df_vente['Date mutation'], format="%d-%Y-%m %H:%M:%S")
df_vente['Date'] = df_vente['Date mutation'].astype('str')
df_vente = df_vente[['id', 'Date', 'Valeur fonciere', 'id']]
vente = df_vente.astype(object).where(pd.notnull(df_vente), None)
vente
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "valeur_fonciere"
)

cursor = db.cursor()

sql = "INSERT INTO vente(id, date_vente, valeur, id_bien) VALUES(%s,%s,%s,%s)"
for i,row in vente.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
