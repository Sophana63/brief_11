```python
import pandas as pd
from pandasql import sqldf
import mysql.connector as mysql
```

```python
df_dep = pd.read_excel('Donnees_Immo/fr-esr-referentiel-geographique.xlsx')
dep = df_dep[['dep_code', 'dep_nom', 'reg_code']]
q = "SELECT * FROM dep GROUP BY dep_code"
dep = sqldf(q, globals())
dep
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "valeur_fonciere"
)

cursor = db.cursor()

sql = "INSERT INTO departement(id, nom_dep, id_reg) VALUES(%s,%s,%s)"
for i,row in dep.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
