```python
import pandas as pd
from pandasql import sqldf
import mysql.connector as mysql
```

```python
df_region = pd.read_excel('Donnees_Immo/fr-esr-referentiel-geographique.xlsx')
region = df_region[['reg_code', 'reg_nom', 'regrgp_nom']]
q = "SELECT * FROM region GROUP BY reg_code"
region = sqldf(q, globals())
#region = region[['reg_code', 'reg_nom', 'regrgp_nom']]
region
```

```python
db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "",
    database = "valeur_fonciere"
)

cursor = db.cursor()

sql = "INSERT INTO region(id, nom_region, groupe) VALUES(%s,%s,%s)"
for i,row in region.iterrows():
    cursor.execute(sql, tuple(row))    
    db.commit()

db.close()
```
